```python
# my analysis
import ROOT
import uproot as up
import numpy as np

f = ROOT.TFile("output.root", "RECREATE")

ROOT.gStyle.SetOptStat(0)
need = 0
current_tree= 0 
amount = 5
    
c1 = ROOT.TCanvas("c1", "5_42")
c2 = ROOT.TCanvas("c2", "5_100")
c3 = ROOT.TCanvas("c3", "10_100")
c4 = ROOT.TCanvas("c4", "10_275")
c5 = ROOT.TCanvas("c5", "18_275")

c1.Divide(2, 2)
c2.Divide(2, 2)
c3.Divide(2, 2)
c4.Divide(2, 2)
c5.Divide(2, 2)

c1.Print("output.pdf[")

def add_text(pad, text, x=0.1, y=0.9):
    pad.cd()
    latex = ROOT.TLatex()
    latex.SetNDC()
    latex.SetTextSize(0.04)
    latex.DrawLatex(x, y, text)
        
for i in range(amount):
    current_tree=current_tree+1
    if current_tree == 1 and need == 0:
        title = "5_41"
        events_tree = up.open("/scratch/kspring/100k_dj/5_41/eicrecon_5_41_all.root")["events"]
    if current_tree == 2 and need == 0:
        title = "5_100" 
        events_tree = up.open("/scratch/kspring/100k_dj/5_100/eicrecon_5_100_all.root")["events"]
    if current_tree == 3 and need == 0:
        title = "10_100"
        events_tree = up.open("/scratch/kspring/100k_dj/10_100/eicrecon_10_100_all.root")["events"]
    if current_tree == 4 and need == 0:
        title = "10_275" 
        events_tree = up.open("/scratch/kspring/100k_dj/10_275/eicrecon_10_275_all.root")["events"]
    if current_tree == 5 and need == 0:
        title = "18_275" 
        events_tree = up.open("/scratch/kspring/100k_dj/18_275/eicrecon_18_275_all.root")["events"]
               
    if current_tree == 1 and need == 1:
        title = "18_275_2k" 
        events_tree = up.open("/scratch/kspring/fromifarm9/kspring/djangoh_output/djangoh.eicrecon_18_275_out.root")["events"] 
    if current_tree == 2 and need == 1: 
        title = "10_275_2k" 
        events_tree = up.open("/scratch/kspring/fromifarm9/kspring/djangoh_output/djangoh.eicrecon_10_275_out.root")["events"] 
    if current_tree == 3 and need == 1: 
        title = "10_100_2k" 
        events_tree = up.open("/scratch/kspring/fromifarm9/kspring/djangoh_output/djangoh.eicrecon_10_100_out.root")["events"] 
    if current_tree == 4 and need == 1: 
        title = "5_100_2k" 
        events_tree = up.open("/scratch/kspring/fromifarm9/kspring/djangoh_output/djangoh.eicrecon_5_100_out.root")["events"]
    if current_tree == 5 and need == 1: 
        title = "5_41_2k" 
        events_tree = up.open("/scratch/kspring/fromifarm9/kspring/djangoh_output/djangoh.eicrecon_5_41_out.root")["events"] 
    if i == 0:
        canvas = c1
    elif i == 1:
        canvas = c2
    elif i == 2:
        canvas = c3
    elif i == 3:
        canvas = c4
    elif i == 4:
        canvas = c5
    if need == 0:
        if i != 0:
            h1_elec =ROOT.TH2F(f"{title}", f"{title} Scattered electron true momentum vs. polar angle", 100, 0, 180, 100, 0, 5.1) 
            h1_prot =ROOT.TH2F(f"{title}", f"{title} Proton true momentum vs. polar angle", 100, 0, 3.5, 100, 40, 41)
            h2_elec =ROOT.TH2F(f"{title}", f"{title} Scattered electron reconstructed momentum vs. polar angle", 100, 0, 170, 100, 0, 5.2)
            h2_prot =ROOT.TH2F(f"{title}", f"{title} Proton reconstructed momentum vs. polar angle", 100, 0, 170, 100, 40, 41)
        else:
            h1_elec =ROOT.TH2F(f"{title}", f"{title} Scattered electron true momentum vs. polar angle", 100, 179.6, 180, 100, 4.999, 5) 
            h1_prot =ROOT.TH2F(f"{title}", f"{title} Proton true momentum vs. polar angle", 100, 1.39, 1.48, 100, 41.0089, 41.011)
            h2_elec =ROOT.TH2F(f"{title}", f"{title} Scattered electron reconstructed momentum vs. polar angle", 100, 179.6, 180, 100, 4.999, 5)
            h2_prot =ROOT.TH2F(f"{title}", f"{title} Proton reconstructed momentum vs. polar angle", 100, 1.39, 1.48, 100, 41.0089, 41.011)
    if need == 1:   
        h1_elec=ROOT.TH2F(f"{title}", f"{title} Scattered electron true momentum vs. polar angle", 100, 0, 170, 100, 0, 5.2)
        h1_prot=ROOT.TH2F(f"{title}", f"{title} Proton true momentum vs. polar angle", 100, 0, 4, 100, 36, 42)
        h2_elec=ROOT.TH2F(f"{title}", f"{title} Scattered electron reconstructed momentum vs. polar angle", 100, 0, 170, 100, 0, 5.2)
        h2_prot=ROOT.TH2F(f"{title}", f"{title} Proton reconstructed momentum vs. polar angle", 100, 0, 170, 100, 36, 46)
        
    gen_status = events_tree["MCParticles.generatorStatus"].array()
    gen_pid = events_tree["MCParticles.PDG"].array() 
    gen_px = events_tree["MCParticles.momentum.x"].array() 
    gen_py = events_tree["MCParticles.momentum.y"].array() 
    gen_pz = events_tree["MCParticles.momentum.z"].array() 
    gen_mass = events_tree["MCParticles.mass"].array() 
    gen_charge = events_tree["MCParticles.charge"].array() 
    gen_vx = events_tree["MCParticles.vertex.x"].array() 
    gen_vy = events_tree["MCParticles.vertex.y"].array() 
    gen_vz = events_tree["MCParticles.vertex.z"].array() 
    
    rec_pid = events_tree["ReconstructedChargedParticles.PDG"].array() 
    rec_px= events_tree["ReconstructedChargedParticles.momentum.x"].array() 
    rec_py= events_tree[ "ReconstructedChargedParticles.momentum.y"].array() 
    rec_pz= events_tree[ "ReconstructedChargedParticles.momentum.z"].array() 
    rec_mass= events_tree["ReconstructedChargedParticles.mass"].array() 
    
    p_beam_true_list = [] 
    e_beam_true_list = [] 
    gen_e_theta_data = [] # Replace with actual theta data in degrees 
    gen_e_momentum_data = [] # Replace with actual momentum data in GeV/c 
    gen_p_theta_data = [] 
    gen_p_momentum_data = [] # 
    rec_e_theta_data = [] # Replace with actual theta data in degrees 
    rec_e_momentum_data = [] # Replace with actual momentum data in GeV/c 
    rec_p_momentum_data = []
    rec_p_theta_data = [] # 
    
    
    for event in range(0, len(events_tree)): # Loop over all events 
        for j in range(0, len(gen_status[event])): # Loop over all thrown particles 
            if gen_status[event][j] == 4 and gen_pid[event][j]==11: 
                e_beam_true = ROOT.Math.LorentzVector('ROOT::Math::PxPyPzM4D<double>')(gen_px[event][j],gen_py[event][j],gen_pz[event][j],gen_mass[event][j]) 
                e_beam_true_list.append(e_beam_true) 
            if gen_status[event][j] == 4 and gen_pid[event][j]==2212: 
                p_beam_true = ROOT.Math.LorentzVector('ROOT::Math::PxPyPzM4D<double>')(gen_px[event][j],gen_py[event][j],gen_pz[event][j],gen_mass[event][j]) 
                p_beam_true_list.append(p_beam_true) 
            if gen_status[event][j]==1: 
                gen_vec = ROOT.Math.LorentzVector('ROOT::Math::PxPyPzM4D<double>')(gen_px[event][j],gen_py[event][j],gen_pz[event][j],gen_mass[event][j]) 
                gen_vertex = ROOT.Math.XYZVector(gen_vx[event][j],gen_vy[event][j],gen_vz[event][j])
                if gen_pid[event][j]==11:
                    gen_vec_theta = gen_vec.Theta() 
                    gen_vec_theta_deg = np.rad2deg(gen_vec_theta) 
                    gen_e_theta_data.append(gen_vec_theta_deg)
                    gen_e_momentum_data.append(gen_vec.P()) 
                    h1_elec.Fill(gen_vec_theta_deg,gen_vec.P()) #look at, may be a (not so worrysum) bug 
                if gen_pid[event][j] == 2212: 
                    gen_vec_theta = gen_vec.Theta() 
                    gen_vec_theta_deg = np.rad2deg(gen_vec_theta)
                    gen_p_theta_data.append(gen_vec_theta_deg) 
                    gen_p_momentum_data.append(gen_vec.P()) 
                    h1_prot.Fill(gen_vec_theta_deg,gen_vec.P())        
        for j in range(0, len(rec_pid[event])): # Loop over all thrown particles
            rec_vec = ROOT.Math.LorentzVector('ROOT::Math::PxPyPzM4D<double>')(rec_px[event][j],rec_py[event][j],rec_pz[event][j],rec_mass[event][j])
            if rec_pid[event][j] ==11:
                #print(rec_pid)
                rec_vec_theta = rec_vec.Theta()
                rec_vec_theta_deg = np.rad2deg(rec_vec_theta)
                rec_e_theta_data.append(rec_vec_theta_deg)
                rec_e_momentum_data.append(rec_vec.P())
                h2_elec.Fill(rec_vec_theta_deg,rec_vec.P())
            if rec_pid[event][j]==2212:
                rec_vec_theta = rec_vec.Theta()
                rec_p_theta_deg = np.rad2deg(rec_vec_theta)
                rec_p_theta_data.append(rec_p_theta_deg)
                rec_p_momentum_data.append(rec_vec.P())
                h2_prot.Fill(rec_vec_theta_deg,rec_vec.P())
    canvas.cd(1)
    h1_elec.Draw("COLZ")
    h1_elec.SetTitle("True Electron Distribution")
    h1_elec.GetXaxis().SetTitle("Theta")
    h1_elec.GetYaxis().SetTitle("Momentum")

    canvas.cd(2)
    h1_prot.Draw("COLZ")
    h1_prot.SetTitle("True Proton Distribution")
    h1_prot.GetXaxis().SetTitle("Theta")
    h1_prot.GetYaxis().SetTitle("Momentum")


    canvas.cd(3)
    h2_elec.Draw("COLZ")
    h2_elec.SetTitle("Rec. Electron Distribution")
    h2_elec.GetXaxis().SetTitle("Theta")
    h2_elec.GetYaxis().SetTitle("Momentum")
    add_text(canvas.cd(3), "Rec. Electron Distribution", 0.2, 0.85)

    canvas.cd(4)
    h2_prot.Draw("COLZ")
    #add_text("Electron Plot", 0.2, 0.85)

    canvas.Print("output.pdf", "Title: " + title)
final_canvas = ROOT.TCanvas("final_canvas",f"{title}")
final_canvas.Print("output.pdf", "Title: 18_275")
final_canvas.Print("output.pdf]")
f.Write()
f.Close()
        
```
